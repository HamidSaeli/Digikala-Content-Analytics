```python
%%writefile app.py

import streamlit as st
import pandas as pd
import plotly.express as px



# --------------------------
# Page Configuration
# --------------------------

st.set_page_config(

    page_title="سامانه هوشمند تحلیل محتوای دیجی‌کالا",

    page_icon="📊",

    layout="wide"

)



# --------------------------
# Load Data
# --------------------------

@st.cache_data

def load_data():

    df = pd.read_csv(

        "data/processed/content_analytics_dataset.csv"

    )

    return df



data = load_data()



# --------------------------
# Header
# --------------------------

st.title(
"📊 سامانه هوشمند تحلیل محتوای دیجی‌کالا"
)


st.markdown(

"""
این داشبورد با هدف تحلیل کیفیت محتوا،
رضایت مشتری و شناسایی فرصت‌های بهبود
با استفاده از تحلیل داده و هوش مصنوعی توسعه یافته است.
"""

)



# --------------------------
# Sidebar
# --------------------------

st.sidebar.title(
"تنظیمات تحلیل"
)



categories = st.sidebar.multiselect(

    "انتخاب دسته‌بندی",

    options=data["Category1"].dropna().unique()

)



filtered=data.copy()



if categories:

    filtered = filtered[

        filtered["Category1"].isin(categories)

    ]



# --------------------------
# KPI Cards
# --------------------------


col1,col2,col3,col4 = st.columns(4)



col1.metric(

    "تعداد محصولات",

    f"{filtered.product_id.nunique():,}"

)



col2.metric(

    "میانگین امتیاز",

    round(filtered.rate.mean(),2)

)



col3.metric(

    "کیفیت محتوا",

    f"{filtered.Content_Completeness.mean()*100:.1f}%"

)



col4.metric(

    "رضایت مشتری",

    f"{filtered.Customer_Satisfaction.mean()*100:.1f}%"

)



# --------------------------
# Tabs
# --------------------------


tab1,tab2,tab3,tab4 = st.tabs(

[
"📈 کیفیت محتوا",
"⭐ رضایت مشتری",
"🤖 تحلیل هوش مصنوعی",
"📥 گزارش‌ها"
]

)



# ==========================
# Tab 1
# ==========================


with tab1:


    st.subheader(
    "تحلیل کیفیت محتوا"
    )


    category_quality=(

        filtered

        .groupby("Category1")

        ["Content_Completeness"]

        .mean()

        .reset_index()

    )


    fig=px.bar(

        category_quality,

        x="Category1",

        y="Content_Completeness",

        title="کیفیت محتوا بر اساس دسته"

    )


    st.plotly_chart(

        fig,

        use_container_width=True

    )



# ==========================
# Tab 2
# ==========================


with tab2:


    st.subheader(

    "تحلیل رضایت مشتری"

    )


    fig=px.histogram(

        filtered,

        x="rate",

        title="Distribution of Ratings"

    )


    st.plotly_chart(

        fig,

        use_container_width=True

    )



# ==========================
# Tab 3
# ==========================


with tab3:


    st.subheader(

    "محصولات نیازمند بررسی"

    )


    risk_products=(

        filtered

        .sort_values(

            "Opportunity_Score",

            ascending=False

        )

        [

        [

        "title_fa",

        "Brand",

        "Category1",

        "Opportunity_Score"

        ]

        ]

        .head(20)

    )


    st.dataframe(

        risk_products,

        use_container_width=True

    )



# ==========================
# Tab 4
# ==========================


with tab4:


    st.subheader(

    "دانلود گزارش"

    )


    csv = filtered.to_csv(

        index=False,

        encoding="utf-8-sig"

    )


    st.download_button(

        label="⬇ دانلود گزارش CSV",

        data=csv,

        file_name="digikala_content_report.csv"

    )
```
