## 7. Missing Value Analysis


Missing data can affect:

- Content Quality Metrics
- Machine Learning Models
- Business Decisions


We identify columns requiring further cleaning.

```python
products_missing = (
    products
    .isnull()
    .sum()
    .sort_values(
        ascending=False
    )
)


products_missing.head(20)
```

```python
comments_missing = (
    comments_sample
    .isnull()
    .sum()
    .sort_values(
        ascending=False
    )
)


comments_missing.head(20)
```

## 8. Duplicate Records


Duplicate data can create:

- Incorrect KPIs
- Biased analysis
- Wrong business conclusions


```python
print(
    "Product duplicates:",
    products.duplicated().sum()
)


print(
    "Comment duplicates:",
    comments_sample.duplicated().sum()
)
```
